```python
#num_epochs = 50
batch_size = 16
learning_rate = 0.0005

input_shape = (384,512,3)
#num_classes = 2
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
training_data = ImageDataGenerator(
    rescale = 1. /255,
    validation_split = 0.2,

    rotation_range = 5,        # 15도 내외로 이미지 회전
    width_shift_range = 0.05,    # 가로 10% 이내 이동
    height_shift_range = 0.05,   # 세로 10% 이내 이동
    shear_range = 0.05,          # 전단 변환 (찌그러뜨리기)
    zoom_range = 0.05,           # 10% 이내 확대/축소
    #horizontal_flip = True,     # 좌우 반전
    fill_mode = 'nearest'       # 변환 시 생기는 빈 공간을 주변 픽셀로 채움
)
test_data = ImageDataGenerator(
    rescale = 1. / 255,
    validation_split = 0.2
)
```

```python
checkpoints = r'./IMAGE/.ipynb_checkpoints'

if os.path.exists(checkpoints):
    shutil.rmtree(checkpoints)

training_generator = training_data.flow_from_directory(
    './IMAGE/plastic_paper',
    batch_size = batch_size,
    target_size = (384,512),
    class_mode = 'categorical',
    shuffle = True,
    subset = 'training'
)
test_generator = test_data.flow_from_directory(
    './IMAGE/plastic_paper',
    batch_size = batch_size,
    target_size = (384,512),
    class_mode = 'categorical',
    shuffle = True,
    subset = 'validation'
)
```

```python
print(training_generator.class_indices)
```

```python
batch_samples = next(iter(training_generator))

print('True value : ', batch_samples[1][0])
plt.imshow(batch_samples[0][0])
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()
model.add(Conv2D(filters = 32, kernel_size = 3, activation = 'relu', input_shape = input_shape))
model.add(MaxPooling2D(pool_size = 2))
model.add(Dropout(0.2))
model.add(Conv2D(filters = 64, kernel_size = 3, activation = 'relu'))
model.add(MaxPooling2D(pool_size = 2))
model.add(Dropout(0.2))
model.add(Conv2D(filters = 128, kernel_size = 3, activation = 'relu')) # 필터 수 128
model.add(MaxPooling2D(pool_size = 2))
model.add(Dropout(0.2))

model.add(Flatten())
model.add(Dense(100, activation = 'relu'))
model.add(Dropout(0.4))
model.add(Dense(2, activation = 'softmax'))
```

```python
early_stopping = EarlyStopping(
    monitor = 'val_loss',
    patience = 5,
    verbose=1,
    mode = 'min',
    restore_best_weights = True
)
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam',
              loss = 'categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
history = model.fit(
    training_generator,
    epochs = 100,
    steps_per_epoch = int(len(training_generator) / batch_size),
    validation_steps = int(len(test_generator) / batch_size),
    validation_data = test_generator,
    verbose = 1,
    callbacks=[early_stopping]
)
```
